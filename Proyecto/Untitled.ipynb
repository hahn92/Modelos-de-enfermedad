```python
import networkx as nx
import matplotlib.pyplot as plt

ORF = 1

nodos = []
vertices = []
vertices_pesos = []
for n in range(3+(ORF-1), len(Cadena1), 3):
#for n in range(3+(ORF-1), 209, 3):
    if Cadena1[n-3:n] in nodos:
        pass
    else:
        nodos.append(Cadena1[n-3:n])
    if n > 3:
        vertices.append([Cadena1[n-6:n-3], Cadena1[n-3:n]])
      
print(nodos)

vert = []
for v in vertices:
    
    if v in vert:
        pass
    else:
        vert.append(v)
        vertices_pesos.append([v[0], v[1], vertices.count(v)])
print(vertices_pesos)
```

```python
DG = nx.DiGraph()
for row in vertices_pesos: 
    DG.add_edge(
        row[0],
        row[1],
        weight=row[2]
    )
DG.nodes(data=True)
```

```python
nx.draw_spectral(
    DG,
    node_color="lightblue",
    edge_color="grey",
    font_size=5,
    width=1,
    with_labels=True,
    node_size=200
)
```

```python
nx.draw_circular(
    DG,
    node_color="lightblue",
    edge_color="grey",
    font_size=5,
    width=1,
    with_labels=True,
    node_size=200
)
```

```python
Gr = nx.DiGraph()
Gr.add_weighted_edges_from(vertices_pesos)
pos = nx.spring_layout(Gr)
list_edges = list(Gr.edges())
last = ()
nx.draw(Gr, pos=pos, with_labels=True, edgelist=list_edges, node_size=600)
edge_labels = dict([((u, v,), d['weight']) for u, v, d in Gr.edges(data=True)])
nx.draw_networkx_edge_labels(Gr, pos=pos, edgelist=list_edges, edge_labels=edge_labels)
plt.axis('off')
plt.show()
```
